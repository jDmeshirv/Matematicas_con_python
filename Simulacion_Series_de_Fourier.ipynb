#Simulación de series de Fourier con Maquina de Euclides
$f(x)=\sum_{n=0}^\infty \frac{4}{(2n+1)\pi}sin[(2n+1)x]$

```
import numpy as np
from matplotlib import pylab
from google.colab import widgets
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
import matplotlib.pyplot as plt
def Cserie_de_fourier(value):
  fig,(ax1,ax2)=plt.subplots(1,2)
  fig.set_size_inches(15,7.5)
  pointsx=[]
  pointsy=[]
  sumax=[]
  sumay=[]
  m=10
  n=np.arange(0,m+1)
  x=np.linspace(0,2*np.pi,1000)
  A=4/np.pi
  A_i=A/(2*n+1)
  for i in n:
    pointx=A_i[i]*np.cos((2*i+1)*value)
    pointy=A_i[i]*np.sin((2*i+1)*value)
    pointsx.append(pointx)
    pointsy.append(pointy)
  for k in n:
    suma_x=np.sum(pointsx[:k])
    sumax.append(suma_x)
    suma_y=np.sum(pointsy[:k])
    sumay.append(suma_y)  
  for i in n:
    ax1.plot(sumax[i]+A_i[i]*np.cos((2*i+1)*x),sumay[i]+A_i[i]*np.sin((2*i+1)*x))
    ax1.plot(sumax[i]+A_i[i]*np.cos((2*i+1)*value),sumay[i]+A_i[i]*np.sin((2*i+1)*value),'o')
  ax1.plot([0,sumax[0]+A_i[0]*np.cos((2*0+1)*value)],[0,sumay[0]+A_i[0]*np.sin((2*0+1)*value)])
  for i in range(m-1):
    ax1.plot([sumax[i]+A_i[i]*np.cos((2*i+1)*value),sumax[i+1]+A_i[i+1]*np.cos((2*(i+1)+1)*value)],[sumay[i]+A_i[i]*np.sin((2*i+1)*value),sumay[i+1]+A_i[i+1]*np.sin((2*(i+1)+1)*value)])
  ax1.plot([sumax[-1]+A_i[-1]*np.cos((2*m+1)*value),20],[sumay[-1]+A_i[-1]*np.sin((2*m+1)*value),sumay[-1]+A_i[-1]*np.sin((2*m+1)*value)],color='orange')
  ax1.grid()
  ax1.set_xlim([-3,3.5])
  ax1.set_ylim([-3,3])
  f=[]
  for i in n:
    f_x=A/(2*i+1)*np.sin((2*i+1)*x)
    f.append(f_x)
  fx=np.array(f)
  h=fx[0]
  for i in range(m-1):
    h=h+fx[i+1]
  ax2.plot(x,h)
  ax2.grid()
  ax2.set_xlim([0,6])
  ax2.set_ylim([-3,3])
  ax2.plot([-20,20],[sumay[-1]+A_i[-1]*np.sin((2*m+1)*value),sumay[-1]+A_i[-1]*np.sin((2*m+1)*value)],color='orange')
  return 

interact(Cserie_de_fourier, value=widgets.FloatSlider(min=0, max=2*np.pi, step=0.1, value=0));
```
